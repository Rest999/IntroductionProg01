### Iris DataSetを使ってsklearnを触ろう


```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#データセット確認
print(iris.DESCR)
```

```python
#:Number of Instances: 150 (50 in each of three classes)
#:Number of Attributes: 4 numeric, predictive attributes and the class
#データフレーム形式にしてデータを扱いやすくする
import pandas as pd

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
#欠損値の確認
df.isnull().sum()
```

```python
#欠損値がなかったため前処理は行わず
#データセットの作成
#75%を訓練データ、25%をテストデータに分割
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris['data'], iris['target'], test_size=0.25,  random_state=0)

# x_trainにデータセットの75%が含まれている
pd.DataFrame(x_train, columns=iris.feature_names)
```

```python
#x_testには残りのデータセットの25%が含まれている
pd.DataFrame(x_test, columns=iris.feature_names)
```

```python
#手法は最近読み始めたデータサイエンスの名書　The elements of Statistical Learning https://link.springer.com/book/10.1007/978-0-387-84858-7 の 2.3.2 Nearest Neighbor Methodsからk個に対応した　k-Nearest Neighbor Methods を使ってみる ※理由は特にないです
from sklearn.neighbors import KNeighborsClassifier

#kを 1-50まで実行してそれぞれのkの数と評価スコアを表示する
#実行数と評価した数値を格納するリストを作成する
list_numOfK = []
list_score = []

for k in range(1,50):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train, y_train)

#予測 x_testを使って学習させる
  Y_pred = knn.predict(x_test)
#学習したものを使って評価スコアを算出
  score = knn.score(x_test, y_test)
  print("[%d] score: {:.2f}".format(score) % k)


  list_numOfK.append(k)
  list_score.append(score)

import matplotlib.pyplot as plt
#k=1-50までの各評価スコアをプロットする。
plt.ylim(0.9, 1.0)
plt.xlabel("n_neighbors")
plt.ylabel("score")
plt.plot(list_numOfK, list_score)
```
